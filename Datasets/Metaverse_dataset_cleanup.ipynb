```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 

import pandas as pd
```

```python
df = pd.read_csv(filepath_or_buffer='./metaverse_transactions_dataset.csv')
df.head(5)
```

```python
mv = pd.DataFrame()
mv = df.drop(['sending_address','receiving_address','ip_prefix'], axis = 1)
mv.head()
```

```python
print(mv.columns)
```

```python
mv['timestamp'] = pd.to_datetime(mv['timestamp'])
mv.head()
```

```python
mv.rename(columns={'timestamp': 'Date'}, inplace=True)
mv
```

```python
mv.isnull().sum()
```

```python
mv.to_csv('../Parsed Datasets/Metaverse.csv', index=False)
```

```python
#Reading data
meta = pd.read_csv(filepath_or_buffer='../Parsed Datasets/meta.csv')
nvdia = pd.read_csv(filepath_or_buffer='../Parsed Datasets/cleaned_NVDA_data.csv')
google = pd.read_csv(filepath_or_buffer='../Parsed Datasets/GoogleParsed.csv')
msft = pd.read_csv(filepath_or_buffer='../Parsed Datasets/MSFT_parsed.csv')
rblx = pd.read_csv(filepath_or_buffer='../Parsed Datasets/rblx.csv')
```

```python
#change datatyep
meta['Date'] = pd.to_datetime(meta['Date'])
nvdia['Date'] = pd.to_datetime(nvdia['Date'])
google['Date'] = pd.to_datetime(google['Date'])
msft['Date'] = pd.to_datetime(msft['Date'])
rblx['Date'] = pd.to_datetime(rblx['Date'])
```

```python
meta['Date'].dtype
```

```python
join_meta_metaverse = pd.merge(mv,meta, how ='inner', on='Date')
```

```python
join_meta_metaverse.isnull().sum()
```

```python
join_meta_metaverse_nvdia = pd.merge(join_meta_metaverse,nvdia, how ='inner', on='Date', suffixes=['_meta','_nvdia'])
join_meta_metaverse_nvdia
```

```python
join_meta_metaverse_nvdia_google = pd.merge(join_meta_metaverse_nvdia,google, how ='inner', on='Date')
join_meta_metaverse_nvdia_google
```

```python
join_meta_metaverse_nvdia_google_msft = pd.merge(join_meta_metaverse_nvdia_google,msft, how ='inner', on='Date', suffixes=['_google','_msft'])
join_meta_metaverse_nvdia_google_msft
```

```python
join_all_stocks = pd.merge(join_meta_metaverse_nvdia_google_msft,rblx, how ='inner', on='Date',suffixes=['_rblx'])
```

```python
last_column = join_all_stocks.columns[-1]

join_all_stocks.rename(columns={last_column: last_column + '_rblx'}, inplace=True)
join_all_stocks
```
